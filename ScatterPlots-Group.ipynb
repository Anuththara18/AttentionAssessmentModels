```python

import pandas as pd
import matplotlib.pyplot as plt

df  = pd.read_csv("A4.csv")

x2 = df[['id']]
y2 = df[['omission_errors']]
plt.scatter(x2, y2, c ="red")

# Sustained Attention, Focused Attention -> 0 *
#cmdf = df.iloc[[1:22],[48:103]]
cmdf  = pd.read_csv("A4.csv")
x = cmdf[['id']]
y = cmdf[['commission_errors']]
plt.scatter(x, y, c ="blue")

plt.xlabel("ID")
plt.ylabel("Blue = CE, Red = OE")
plt.title("Commission & Omission Errors")

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df  = pd.read_csv("A4.csv")

df.plot(kind='scatter',x='id',y='percentage_no_of_correct_responses', c ="orange", title="Percentage No. of Correct Responses")

# Mean Reaction Time
mrtdf = df.loc[df['mean_reaction_time'] != 0]
mrtdf.plot(kind='scatter',x='id',y='mean_reaction_time', c ="green", title="Mean Reaction Time") 
```

```python
# Total Duration 
# Alternating Attention, Divided Attention -> Not Assessed 
df  = pd.read_csv("A4.csv")
tddf = df.iloc[20:109]
tddf.plot(kind='scatter',x='id',y='total_duration', c ="purple", title="Total Duration")
```
