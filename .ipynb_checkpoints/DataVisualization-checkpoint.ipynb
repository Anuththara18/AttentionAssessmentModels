## Focused Attention
>__Test Variables__ 
>1. Age
>2. % No Of Correct Responses
>3. Commission Errors
>4. Mean Reaction Time

__Graphs__

1. Variation of the No of Correct Responses as a Percentage
***
2. Variation of the Commission Errors
***
3. Variation of the Mean Reaction Time

```python
import pandas as pd
from matplotlib import pyplot as plt

plt.rcParams["figure.figsize"] = [15.00, 7.00]
plt.rcParams["figure.autolayout"] = True

columns = ['%NoOfCorrectResponses', 'Age']

df = pd.read_csv("FocusedAttention.csv", usecols=columns)

df.plot()

plt.title('Variation of the No. of Correct Responses')
plt.xlabel('Child')
plt.ylabel('Value')

plt.show()
```
